# Network Analysis for the **Industries**

```python
#Imports 
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python

G = nx.from_pandas_edgelist(data, source='symbol', target='sector')


nx.draw(G, with_labels=True)
plt.show()

# Get the number of nodes and edges
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
plt.show()
print(f'Number of nodes: {num_nodes}')
print(f'Number of edges: {num_edges}')
```

It seems that there is a few clusters of industries.

```python

# Test 5 different parameter combinations
parameter_combinations = [
    {'k': 0.1, 'iterations': 2, 'scale' : 10.0 , 'center':(1, 1) },
    {'k': 0.1, 'iterations': 100 ,'scale' : 20.0 , 'center':(2, 2)},
    {'k': 0.5, 'iterations': 50, 'scale' : 30.0 , 'center':(3, 3)},
    {'k': 0.5, 'iterations': 100, 'scale' : 40.0 , 'center':(4, 4)},
    {'k': 1, 'iterations': 50, 'scale' : 15.0 , 'center':(0.5, 0.5)},
]

# Compute the Fruchterman-Reingold layout for each combination
for i, params in enumerate(parameter_combinations):
    pos = nx.fruchterman_reingold_layout(G, **params)
    
    # Calculate the mean squared displacement of the nodes
    displacement = 0
    for node, (x, y) in pos.items():
        displacement += (x - params['center'][0])**2 + (y - params['center'][1])**2
    displacement /= len(pos)
    
    # Print the parameter combination and the mean squared displacement
    print(f'Parameter combination {i+1}: {params}')
    print(f'Mean squared displacement: {displacement:.2f}')
```

The Fruchterman-Reingold algorithm, a widely used graph layout algorithm, allows us to visually represent the stock market network in a visually appealing manner by determining the optimal positioning of nodes in the graph. We aim to obtain the best results using this algorithm by optimizing the Mean Squared Displacement (MSD) metric, which measures the average distance of nodes from the center of the graph layout and allows for effective comparison of different layout options.

```python
# The chosen paramters are: 
params = {'k': 0.1, 'iterations': 2, 'scale' : 10.0 , 'center':(1, 1) }

# Compute the Fruchterman-Reingold layout
pos = nx.fruchterman_reingold_layout(G, **params)

# Plot the graph
plt.figure()
nx.draw_networkx(G, pos)

# Show the plot
plt.show()
```

**The chosen parameter combination for the Fruchterman-Reingold layout was:  {'k': 0.1, 'iterations': 2, 'scale' : 10.0 , 'center':(1, 1) }.**



**Based on the obtained results, the Fruchterman-Reingold algorithm was applied to arrange the nodes in the graph in an aesthetically pleasing way. To improve the layout, a combination of parameters was selected, including 'k': 0.1, 'iterations': 2, and 'temperature': 0.01, which produced the most appealing and readable layout for the given network.**
